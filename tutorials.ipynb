### Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

**Basic Graph ðŸ“ˆ**

```python
# Data
x = np.array([0, 1, 2, 3])
y = x * 2
y
```

```python
# Resizing the graph
plt.figure(figsize=(5, 3), dpi=150)

# Ploting the graph
plt.plot(x, y, 
         label='2x', 
         linestyle = "--", linewidth = 2, color = '#7B0323', # work with the line 
         marker = ".", markersize = 12, markeredgecolor = 'black' # working the dots (i.e. markers)
        )
# shorthand notation for the same thing
# plt.plot(x, y, 'r.--', label='2x')


# 2nd line on the graph
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2, x2**2, 
         label = "x^2",
         linewidth=2,
         color='#4d004d'
        )


# adding ticks(i.e. intervals in the number line)
plt.xticks([0, 1, 2, 3])
plt.yticks(np.arange(0, (x2[-1]**2 + 1), 2))

# adding lebels
plt.title("Graph: 2x & x^2", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 14})
plt.xlabel("x-axis", fontdict={"fontname": "Comic Sans MS"})
plt.ylabel("y-axis", fontdict={"fontname": "Comic Sans MS"})

# adding legend
plt.legend()

# Save the graph
plt.savefig('myfirstgraph.png', dpi=300)

plt.show()  # optional, to avaid annoying text above the graph
```

**Bar Charts ðŸ“Š**

```python
# resizing the figure
plt.figure(figsize=(5, 3), dpi=150)

labels = ['A', 'B', 'C']
values = [1, 4, 2]

# Plotting the bar chart
bar_chart = plt.bar(labels, values)

# beutifying the chart
# bar_chart[0].set_hatch('-')
# bar_chart[1].set_hatch('|/')
# bar_chart[2].set_hatch('|')
pattern = ['-', '|/', '|']
for bar in bar_chart:
    bar.set_hatch(pattern.pop(0))

plt.show()
```

```python
# Loading Data
gas = pd.read_csv('gas_prices.csv').drop(columns=["Unnamed: 0"])
gas.tail(5)
```

```python
countires = gas.columns.to_list()[1:] # getting required column names to itirate on for the graph generation

# Configuring the graph size
plt.figure(figsize=(10, 7), dpi=120)

# Ploting the data
for country in countires:
    plt.plot(gas.Year, gas[f"{country}"], label=f"{country}", marker=".")

# Adding labels
plt.title("Gas price over time (in USD)")
plt.xlabel("Year")
plt.ylabel("Price (in USD)")

# Ticks
plt.xticks(gas.Year[::3])

plt.legend()

plt.show()
```
