### Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

**Basic Graph ðŸ“ˆ**

```python
# Data
x = np.array([0, 1, 2, 3])
y = x * 2
y
```

```python
# Resizing the graph
plt.figure(figsize=(5, 3), dpi=150)

# Ploting the graph
plt.plot(x, y, 
         label='2x', 
         linestyle = "--", linewidth = 2, color = '#7B0323', # work with the line 
         marker = ".", markersize = 12, markeredgecolor = 'black' # working the dots (i.e. markers)
        )
# shorthand notation for the same thing
# plt.plot(x, y, 'r.--', label='2x')


# 2nd line on the graph
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2, x2**2, 
         label = "x^2",
         linewidth=2,
         color='#4d004d'
        )


# adding ticks(i.e. intervals in the number line)
plt.xticks([0, 1, 2, 3])
plt.yticks(np.arange(0, (x2[-1]**2 + 1), 2))

# adding lebels
plt.title("Graph: 2x & x^2", fontdict={'fontname': 'Comic Sans MS', 'fontsize': 14})
plt.xlabel("x-axis", fontdict={"fontname": "Comic Sans MS"})
plt.ylabel("y-axis", fontdict={"fontname": "Comic Sans MS"})

# adding legend
plt.legend()

# Save the graph
plt.savefig('myfirstgraph.png', dpi=300)

plt.show()  # optional, to avaid annoying text above the graph
```

**Bar Charts ðŸ“Š**

```python
# resizing the figure
plt.figure(figsize=(5, 3), dpi=150)

labels = ['A', 'B', 'C']
values = [1, 4, 2]

# Plotting the bar chart
bar_chart = plt.bar(labels, values)

# beutifying the chart
# bar_chart[0].set_hatch('-')
# bar_chart[1].set_hatch('|/')
# bar_chart[2].set_hatch('|')
pattern = ['-', '|/', '|']
for bar in bar_chart:
    bar.set_hatch(pattern.pop(0))

plt.show()
```

```python
# Loading Data
gas = pd.read_csv('gas_prices.csv').drop(columns=["Unnamed: 0"])
gas.tail(5)
```

```python
# Configuring the graph size
plt.figure(figsize=(8.5, 7), dpi=120)

# Ploting the data
countires = gas.columns.to_list()[1:] # getting required column names to itirate on for the graph generation
for country in countires:
    plt.plot(gas.Year, gas[f"{country}"], label=f"{country}", marker=".")

# Adding labels
plt.title("Gas price over time (in USD)", fontdict={"fontweight" : "bold", "fontsize": 14})
plt.xlabel("Year")
plt.ylabel("Price (in USD)")

# Ticks
plt.xticks(gas.Year[::3])

plt.legend()

plt.savefig("Gas_price_vs_year.png", dpi=300)

plt.show()
```

```python
# Loading Data
fifa = pd.read_csv("fifa_data.csv").drop(columns=["Unnamed: 0.1", "Unnamed: 0"])
fifa.sample(5)
```

**Histograms**

```python
bins = np.arange(40, 101, 10)

plt.hist(fifa.Overall, bins=bins, color="#7926d1")

plt.xticks(bins)

plt.ylabel("Number of players")
plt.xlabel("Skill Level")
plt.title("Distribution of Player Skills in FIFA 2018", fontdict={"fontweight":"bold"})

plt.show()
```

**Pie Chart**

```python
left_foot = fifa.loc[fifa["Preferred Foot"] == "Left"].shape[0]
right_foot = fifa.loc[fifa["Preferred Foot"] == "Right"].count().iloc[0]

labels = ['Left', 'Right']
colors = ["red", "#05f7d7"]

plt.pie([left_foot, right_foot], labels= labels, colors=colors, autopct="%.2f%%")
plt.title("Foot Prefernce of FIFA players")
plt.show()
```

Specifically, Series.__getitem__ will no longer support positional indexing; instead, it will treat keys as labels.

Solution: To ensure compatibility with future versions of pandas, you should explicitly access the count using .iloc[0] (if you want to use .count()) or .shape[0].\

```python
## Ploting a Pie char for the Weight column Data

# Converting all data (string, non-str (i.e. NaN, int)) into string and then replacing the "lbs" part of each data with ""
fifa['Weight'] = fifa['Weight'].astype(str).str.replace('lbs', '', regex=True)

fifa['Weight'] = pd.to_numeric(fifa['Weight'], errors='coerce')  # Convert to numbers, invalid ones become NaN (errors='coerce' is for that)

# Alternatively, 
# fifa.Weight = [int(x.strip("lbs")) if type(x) == str else x for x in fifa.Weight]  # I didn't get this

light = fifa.loc[fifa.Weight < 125].count().loc["Name"]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count().iloc[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].shape[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].shape[0]
heavy = fifa.loc[fifa.Weight > 200].shape[0]
```

```python
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']

explode = (0.4, 0.01, 0.01, 0.009, 0.4)
plt.style.use('ggplot') # for better color scheme
plt.pie(weights, labels=labels, autopct="%.2f%%", pctdistance = 0.9, explode=explode)

plt.title("weight distribution of FIFA players".title() + " (in lbs)")

plt.show()
```

```python
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]["Overall"]
madrid = fifa.loc[fifa.Club == "Real Madrid"]["Overall"]
revs = fifa.loc[fifa.Club == "New England Revolution"]["Overall"]

labels = ["FC Barcelona", "Real Madrid", "NE Revolution"]

plt.figure(figsize=(4, 7))

boxes = plt.boxplot([barcelona, madrid, revs], tick_labels=labels, 
                    patch_artist=True, 
                    medianprops={"linewidth": 2}) # patch_artist=True to be able to change the face style of the boxes (e.g. facecolor = "<color>")
 
for box in boxes['boxes']:
    # set edge color of the boxes
    box.set(color="#4286f4", linewidth=2)

    # Change fill color
    box.set(facecolor = "#e0e0e0")

plt.title("Professonal Football Team Comparison")
plt.ylabel("FIFA Overall Rating")

plt.style.use("default")

plt.show()
```
